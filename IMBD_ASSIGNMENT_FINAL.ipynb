## Q1

```python
%%time
def grader_1(q1):
    q1_results  = pd.read_sql_query(q1,conn)
    print(q1_results.head(10))
    assert (q1_results.shape == (232,3))
query1 = """select p.name, m.title, CAST(SUBSTR(TRIM(m.year),-4) AS INTEGER) as year
from genre as g 
inner join m_genre as mg
on mg.gid=g.gid 
inner join movie as m 
on mg.mid=m.mid
inner join 
m_director as md 
on m.mid=md.mid
inner join person as p
on p.pid=md.pid
where g.name like '%comedy%' 
and
(CAST(SUBSTR(TRIM(m.year),-4) AS INTEGER)%4=0)
and
((CAST(SUBSTR(TRIM(m.year),-4) AS INTEGER)%100!=0) or (CAST(SUBSTR(TRIM(m.year),-4) AS INTEGER)%100=0) and (CAST(SUBSTR(TRIM(m.year),-4) AS INTEGER)%400=0))"""
grader_1(query1)
```

# Q3

```python
%%time
def grader_3(q3):
    q3_results  = pd.read_sql_query(q3,conn)
    print(q3_results.head(10))
    assert (q3_results.shape == (300,1))

query3 ="""
with 
cast_before_1970 as
(Select p.PID from Person p 
inner join 
(
    select trim(mc.PID) PD, mc.MID from M_cast mc 
where mc.MID 
in 
(
    select mv.MID from Movie mv where CAST(SUBSTR(mv.year,-4) AS Integer)<1970
)
) r1 
on r1.PD=p.PID ),
cast_after_1990 as
(Select p.PID from Person p 
inner join 
(
    select trim(mc.PID) PD, mc.MID from M_cast mc 
where mc.MID 
in 
(
    select mv.MID from Movie mv where CAST(SUBSTR(mv.year,-4) AS Integer)>1990
)
) r1 
on r1.PD=p.PID)

select p.name as "Actor_Name"
from person as p
where p.pid in cast_before_1970 and trim(p.pid) in cast_after_1990
"""
grader_3(query3)
```

# Q5

```python
%%time
def grader_5a(q5a):
    q5a_results  = pd.read_sql_query(q5a,conn)
    print(q5a_results.head(10))
    assert (q5a_results.shape == (4,2))

query5a="""select m.year, count(*) as movie_count from movie m
inner join
(select distinct mid from M_Cast  
where mid not in                
(select mc.mid from M_Cast mc 
join Person p on p.pid = trim(mc.pid)
where p.gender = 'Male')
) as view on view.mid = m.mid
group by m.year"""
grader_5a(query5a)
```

```python
%%time
def grader_5b(q5b):
    q5b_results  = pd.read_sql_query(q5b,conn)
    print(q5b_results.head(10))
    assert (q5b_results.shape == (4,3))

query5b = """
with 
female_movies as (select m.year as year, count(*) as female_count from movie m
inner join
(select distinct mid from M_Cast  
where mid not in                
(select mc.mid from M_Cast mc 
join Person p on p.pid = trim(mc.pid)
where p.gender = 'Male')
) as view on view.mid = m.mid
group by m.year),
total_movies as (select CAST(SUBSTR(movie.year,-4) AS Integer) year, count(*) as total_movies_count
from movie 
group by CAST(SUBSTR(movie.year,-4) AS Integer) )

select fm.year,(round(fm.female_count,2))/(round(tm.total_movies_count,2)) as Percentage_Female_Only_Movie,tm.total_movies_count as total_movies
from female_movies as fm
inner join total_movies as tm
on fm.year=tm.year
"""
grader_5b(query5b)
```

```python
%%time
def grader_6(q6):
    q6_results  = pd.read_sql_query(q6,conn)
    print(q6_results.head(10))
    assert (q6_results.shape == (3473, 2))

query6 ="""
with movie_actors as
(select count(*) as count,mid
from m_cast as mc
group by mid
order by count desc)
select m.title, ma.count
from movie as m
inner join movie_actors as ma
on m.mid=ma.mid
order by count desc"""
grader_6(query6)
```

# Q7

```python
%%time
def grader_7(q7):
    q7_results  = pd.read_sql_query(q7,conn)
    print(q7_results.head(10))
    assert (q7_results.shape == (1, 2))

query7 ="""
with 
distinct_year as 
(select distinct CAST(SUBSTR(movie.year,-4) AS Integer) as year from movie)

select count(*) as count,dy.year as start_of_decade 
from distinct_year as dy
inner join movie as m
on m.year>=dy.year
and m.year<=(dy.year+9)
group by start_of_decade
order by count desc
limit 1
"""
grader_7(query7)
```

```python
%%time

def grader_8(q8):
    q8_results  = pd.read_sql_query(q8,conn)
    print(q8_results.head(10))
    print(q8_results.shape)
    assert (q8_results.shape == (245, 2))

query8 = """select p.name,view.count 
from (select mc.pid as actor_pid,md.pid as director_pid,count(mc.mid) as count   
from m_cast as mc
inner join m_director md 
on md.mid=mc.mid              
group by mc.pid ,md.pid) as view
inner join person p 
on view.actor_pid=p.pid
where view.count = (select count(*) as count   
                     from m_cast as mc
                     join m_director md 
                    on md.mid=mc.mid
                     where mc.pid=view.actor_pid 
                     group by mc.pid,md.pid 
                     order by count(*) desc
                     limit 1)
                and view.director_pid = (select pid 
                               from person 
                               where name like '%Yash Chopra%'
                              )
            order by view.count desc"""
grader_8(query8)
```
